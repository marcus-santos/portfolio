## Data Exploration

### Correlations

We set the starting year $ y_0 $ and the last year $ y_n $ and calculate de log returns $ ln(t_n) - ln(t_{n-1}) $

```python
y_0, y_1 = '2011', '2021'
tickers = ['IMA-B', 'IFIX', 'IBOV', 'BDRX', 'DOLOF', 'NASDAQ', 'S&P 500', 'DJIA', 'WTICO', 'EURUSD']
data = df[y_0:y_1][tickers].copy()
data.loc[:, 'NASDAQ':] = data.loc[:, 'NASDAQ':].mul(data['DOLOF'], axis=0)
lreturns = np.log( data ) - np.log( data.shift(1) )
lreturns
```

```python
sns.heatmap(lreturns.corr(), annot=True, fmt='.2f');
```

We find strong correlations between `NASDAQ`, `S&P 500` and `DJIA`

Also, `EURUSD` shows correlation above `0.8` against `DOLOF`

Dropping `NASDAQ`, `DJIA` and `EURUSD`

```python
lreturns.drop(columns=['DJIA', 'NASDAQ', 'EURUSD'], inplace=True)
sns.heatmap(lreturns.corr(), annot=True, fmt='.2f');
```

## Basic Markowitz

We use `PyPortfolioOpt`to make a basic Markowitz portfolio with the Mean-Variance Optimization

```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
from pypfopt import plotting

# Calculate expected returns and sample covariance
mu = expected_returns.mean_historical_return(lreturns, returns_data=True)
S = risk_models.sample_cov(lreturns, returns_data=True)
```

### Expected Returns

```python
mu.plot.barh(figsize=(10,6));
```

### Estimated Variance

```python
sns.heatmap(S, annot=True, fmt='.2f')
```

### Optimize with Max Sharpe

Full date range

```python
ef = EfficientFrontier(mu, S)
weights = ef.max_sharpe()
ex_return, ex_vol, sharpe = ef.portfolio_performance(verbose=True)
```

```python
pd.Series(weights, name="Weights % for Max Sharpe").plot.pie(figsize=(6,6), autopct="%.2f", fontsize=14);
```

## Markowitz Portfolio 3 years for 1 year

```python
def calc_markowitz(returns, verbose=False):
    mu = expected_returns.mean_historical_return(returns, returns_data=True)
    S = risk_models.sample_cov(returns, returns_data=True)
    ef = EfficientFrontier(mu, S)
    weights = ef.max_sharpe()
    r, v, s = ef.portfolio_performance(verbose=verbose)
    return dict(expected_return=r, expected_volatility=v, sharpe=s, weights=weights)


def calc_future_perf(returns, weights, year_begin, year_last=None):
    year_last = year_begin if not year_last else year_last
    future = returns[str(year_begin):str(year_last)].add(1).cumprod(axis=0).sub(1);
    weighted_future = future * pd.Series(weights)
    total = weighted_future.sum(axis=1)
    r = total.iloc[-1]
    v = total.std()
    high, low = total.max(), total.min()
    return dict(real_return = r, 
                real_volatility = v, 
                real_sharpe = (r - 0.02) /  v, 
                real_high=high, real_low=low)

    
def calc_markowitz_year(returns, n_years = 3, n_years_test = 1):
    years = returns.index.year.unique()
    out = []
    for year in years:
        year_last = year + n_years - 1
        year_test = year_last + n_years_test
        if year_test not in years:
            break;
        data = calc_markowitz(returns[str(year):str(year_last)], verbose=False)
        w = data['weights']
        del data['weights']
        data = dict(year_begin=year, year_last=year_last, year_test=year_test, **data)
        fdata = calc_future_perf(returns, w, year_begin=year_test, year_last=year_test)
        
        data = dict(**data, **fdata, **w)
        out.append(data)

    return out

data = calc_markowitz_year(lreturns, n_years = 2, n_years_test = 1)
results = pd.DataFrame(data)
results
```

```python
results.set_index('year_test').loc[: , ['sharpe', 'real_sharpe']].plot.bar(figsize=(10,6));
```

```python
results.set_index('year_test').loc[: , ['expected_return', 'real_return']].add(1).cumprod(axis=0).sub(1).plot(figsize=(10,6));
```

```python
data2 = calc_markowitz(lreturns['2011':'2013'], verbose=False)
w = data2['weights']
del data2['weights']
future = lreturns['2014':'2014'].add(1).cumprod(axis=0).sub(1);

future.plot(figsize=(20,5));
```
